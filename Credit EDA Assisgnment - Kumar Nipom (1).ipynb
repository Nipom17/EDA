```python
# Making a left join because we need all the rows in application data, by making this left join we get historical application data for each applicant.

df_merge=pd.merge(df, df_prev, how='left', on='SK_ID_CURR', suffixes=('_Current', '_Previous'))
df_merge.head()
```

```python
# Univariate Categorical Analysis

cat_colm=['NAME_CONTRACT_TYPE_Current','NAME_CONTRACT_TYPE_Previous',
         'WEEKDAY_APPR_PROCESS_START_Current', 'WEEKDAY_APPR_PROCESS_START_Previous',
         'NAME_YIELD_GROUP','NAME_CLIENT_TYPE']


plt.figure(figsize=(22,25))
for i in (enumerate(cat_colm)):
    plt.subplot(len(cat_colm)//2,2,i[0]+1)
    sns.countplot(x=i[1],hue='NAME_CONTRACT_STATUS',data=df_merge)

plt.show()
```

1. Cash loans and Consumer loans are approved most
2. Previous canceled applications are mostly the cash loans

```python
# Univariate Numerical analysis

cont_colm=['AMT_CREDIT_Previous','AMT_CREDIT_Current','AMT_ANNUITY_Current','AMT_ANNUITY_Previous',
           'CNT_CHILDREN','HOUR_APPR_PROCESS_START_Previous','HOUR_APPR_PROCESS_START_Current']

plt.figure(figsize=(22,25))
for i in (enumerate(cont_colm)):
    plt.subplot(len(cont_colm)//2,3,i[0]+1)
    sns.distplot(df_merge.loc[df_merge.NAME_CONTRACT_STATUS=='Approved',:][i[1]].dropna(),hist=False,label='Approved')
    sns.distplot(df_merge.loc[df_merge.NAME_CONTRACT_STATUS=='Canceled',:][i[1]].dropna(),hist=False,label='Canceled',kde_kws={'bw':0.1})
    sns.distplot(df_merge.loc[df_merge.NAME_CONTRACT_STATUS=='Refused',:][i[1]].dropna(),hist=False,label='Refused',kde_kws={'bw':0.1})
    # we added kde_kws={'bw':0.1} in parameter to overcome bandwidth limitation.
    sns.distplot(df_merge.loc[df_merge.NAME_CONTRACT_STATUS=='Unused offer',:][i[1]].dropna(),hist=False,label='Unused offer')

plt.show() 
```

1.Family with fewer members tends to take more loans.
2.Busiest hours at the bank is from 9am till 2pm and most applications are applied during this period.

```python
# Bivariate Categorical Analysis 
## Creating a table for columns NAME_CONTRACT_STATUS vs NAME_INCOME_TYPE

table = pd.crosstab(index=df_merge['NAME_CONTRACT_STATUS'],columns=df_merge['NAME_INCOME_TYPE'])
table = table.unstack().reset_index(name='counts')

# Plotting barplot

plt.figure(figsize=(8,6))
sns.barplot(data=table.unstack(), x='NAME_CONTRACT_STATUS', y='counts', hue='NAME_INCOME_TYPE')
plt.show()
```

Working category has the highest number of approval ratings

```python
# Bivariate Numerical Analysis
## Plotting subplots for

cont_colm=['AMT_CREDIT_Previous','AMT_CREDIT_Current','AMT_ANNUITY_Current','AMT_ANNUITY_Previous',
           'CNT_CHILDREN','HOUR_APPR_PROCESS_START_Previous','HOUR_APPR_PROCESS_START_Current']
    
plt.figure(figsize=(22,25))

for i in (enumerate(cont_colm)):
    plt.subplot(len(cont_colm)//2,3,i[0]+1)
    sns.boxplot(x='NAME_CONTRACT_STATUS',y=df_merge[i[1]].dropna(),data=df_merge)
plt.show()
```
